```python
reshaped_data=df['Extracurricular Activities'].values.reshape(-1,1)
```

```python
reshaped_data
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit_transform(reshaped_data)
```

```python
encoder.fit_transform(reshaped_data).toarray()
```

```python
encoder.categories_[0]
```

```python
df[encoder.categories_[0]]=encoder.fit_transform(reshaped_data).toarray()
```

```python
df.head()
```

```python
df.drop('Extracurricular Activities',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Performance Index', axis=1)
y = df['Performance Index']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print('Training_accuracy:',r2_score(y_train,reg.predict(x_train)))
```

```python
print("Test_accuracy:",r2_score(y_test,reg.predict(x_test)))
```

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import  r2_score
```

```python
knn_reg=KNeighborsRegressor(n_neighbors=5)
```

```python
knn_reg.fit(x_train,y_train)
```

```python
print('Training_accuracy:',r2_score(y_train,knn_reg.predict(x_train)))
```

```python
print("Test_accuracy:",r2_score(y_test,knn_reg.predict(x_test)))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model= DecisionTreeRegressor(max_depth=4)
```

```python
model.fit(x_train,y_train)
```

```python
print('Training_accuracy:',r2_score(y_train,model.predict(x_train)))
```
